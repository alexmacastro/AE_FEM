## Base de polinomios en dos dimensiones. 

En lo que sigue se considerará dos tipos de bases polinómicas: 

+ Polinomios de grado total $m$. Son polinomios cuya base está generada por productos $x^a\,y^b$, con $a+b \leq m$. Las potencias corresponden con las de un desarrollo en serie de Taylor de una función de 2 variables hasta orden $m$. 

Ejemplo: Para $m=3$, se tomará la siguiente base, de 10 elementos. 

\begin{equation}
\mathcal{B} = \lbrace 1,x,y,x^2,x\,y,y^2,x^3,x^2\,y,x\,y^2,y^3 \rbrace 
\end{equation}

Nótese que la suma de los exponentes es estrictamente menor o igual a 3. 

+ Polinomios de grado parcial $m$. Son polinomios que surgen del producto de dos bases $P^m \times P^m$. Las funciones de la base son de la forma $x^a\,y^b$, con $a\leq m$, $b \leq m$. 

Ejemplo: Para $m=3$ se tomará el producto de $[1,x,x^2,x^3]\times[1,y,y^2,y^3]$, generándose la base de 16 elementos: 

\begin{equation}
\mathcal{B} = \lbrace 1,y,y^2,y^3,x,x\,y,x\,y^2,x\,y^3,x^2,x^2\,y,x^2\,y^2,x^2\,y^3,x^3,x^3\,y,x^3\,y^2,x^3\,y^3 \rbrace 
\end{equation}

Estos polinomios se generan por producto de dos bases unidimensionales.

## Elemento isoparamétrico de 4 nodos bilineal

Este elemento está formado por cuatro funciones de base, $N_1$,$N_2$, $N_3$, $N_4$. 

Las condiciones se establecen en las coordenadas naturales $t,p$.

- 4 puntos, $(-1,-1), (1,-1), (1,1), (-1,1)$. 

- Base $\mathcal{B} = \lbrace {1,t,p,t\,p} \rbrace$

```python
from sympy import init_session
init_session()
```

```python
t,p = symbols('t p')
base = Matrix([[1,t,p,t*p]])
base
```

```python
matM = Matrix.vstack(base.subs([(t,-1),(p,-1)]),
                       base.subs([(t, 1),(p,-1)]),
                       base.subs([(t, 1),(p, 1)]),
                       base.subs([(t, -1),(p, 1)]))
matM
```

```python
Inv_matM = matM.inv()
Inv_matM
```

```python
N1 = base.dot(Inv_matM.col(0))
N2 = base.dot(Inv_matM.col(1))
N3 = base.dot(Inv_matM.col(2))
N4 = base.dot(Inv_matM.col(3))
```

## Elemento isoparamétrico cuadrático

Este elemento está formado por 9 funciones de base, $N_1,N_2,N_3,N_4,N_5,N_6,N_7,N_8,N_9$.  

Las condiciones se establecen en las coordenadas naturales $t,p$.

- 9 puntos, $[-1,0,1]\times[-1,0,1]$

- Base $\mathcal{B} = \lbrace 1,t,t^2 \rbrace \times \lbrace 1,p\,p^2 \rbrace$

```python
base = Matrix([[1,t,t**2,p,t*p,t**2*p,p**2,t*p**2,t**2*p**2]])
base
```

```python
matM = Matrix.vstack(base.subs([(t,-1),(p,-1)]),
                       base.subs([(t, 0),(p,-1)]),
                       base.subs([(t, 1),(p, -1)]),
                       base.subs([(t, 1),(p, 0)]),
                       base.subs([(t, 1),(p, 1)]),
                       base.subs([(t, 0),(p, 1)]),
                       base.subs([(t, -1),(p, 1)]),
                       base.subs([(t, -1),(p, 0)]),
                       base.subs([(t, 0),(p, 0)]))
matM
```

```python
Inv_matM = matM.inv()
Inv_matM
```

```python
N1 = base.dot(Inv_matM.col(0))
N2 = base.dot(Inv_matM.col(1))
N3 = base.dot(Inv_matM.col(2))
N4 = base.dot(Inv_matM.col(3))
N5 = base.dot(Inv_matM.col(4))
N6 = base.dot(Inv_matM.col(5))
N7 = base.dot(Inv_matM.col(6))
N8 = base.dot(Inv_matM.col(7))
N9 = base.dot(Inv_matM.col(8))
```
