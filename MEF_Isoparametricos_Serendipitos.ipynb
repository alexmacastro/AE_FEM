## Elemento serendípito cuadrático de 8 nodos

Este elemento está formado por ocho funciones de base.

Las condiciones se establecen en las coordenadas naturales $t,p$.

- 8 puntos, 

- Base $\mathcal{B} = \lbrace 1,t,t^2 \rbrace \times \lbrace 1,p\,p^2 \rbrace - t^2\,p^2$

```python
from sympy import init_session
init_session()
```

```python
t,p = symbols('t p')
base = Matrix([[1,t,t**2,p,t*p,t**2*p,p**2,t*p**2]])
base
```

```python
matM = Matrix.vstack(base.subs([(t,-1),(p,-1)]),
                       base.subs([(t, 0),(p,-1)]),
                       base.subs([(t, 1),(p, -1)]),
                       base.subs([(t, 1),(p, 0)]),
                       base.subs([(t, 1),(p, 1)]),
                       base.subs([(t, 0),(p, 1)]),
                       base.subs([(t, -1),(p, 1)]),
                       base.subs([(t, -1),(p, 0)]))
matM
```

```python
Inv_matM = matM.inv()
Inv_matM
```

```python
N1 = base.dot(Inv_matM.col(0))
N2 = base.dot(Inv_matM.col(1))
N3 = base.dot(Inv_matM.col(2))
N4 = base.dot(Inv_matM.col(3))
N5 = base.dot(Inv_matM.col(4))
N6 = base.dot(Inv_matM.col(5))
N7 = base.dot(Inv_matM.col(6))
N8 = base.dot(Inv_matM.col(7))
```
